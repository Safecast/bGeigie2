bGeigie3 Endurance Test
=======================

This file analyze the result from the bGeigie3 endurance test.

First we need to open the LOG file and load the data into pandas.

```python
%pylab inline
figsize(16,8)

import pandas as pd

# First we need to split the lines in different files according to line format
with open('301-0814.LOG') as inFile, open('BNXRDD.txt', 'w') as rddFile, open('BNXSTS.txt', 'w') as stsFile:
    for line in inFile:
        if line.startswith('$BNXRDD'):
            rddFile.write(line)
        elif line.startswith('$BNXSTS'):
            stsFile.write(line)
            
# Read RDD file
rdd_col_names = ['Label','ID','Date','CPM','CP5S','TotalCount','RDD Flag','Lat','NS','Lon','EW','alt','GPS Flag','hdop','qu/chksum']
rdd = pd.read_csv('BNXRDD.txt', parse_dates=True, index_col=2,names=rdd_col_names)

# Read STS file
sts_col_names = ['Label','ID','Date','Lon','NS','Lon','EW','# Sat','version','temp','hum','bat','HVPS','SD1','SD2','SD3/chksum']
sts = pd.read_csv('BNXSTS.txt', parse_dates=True, index_col=2, names=sts_col_names)
```

Battery
-------

The battery voltage is also monitored by the board. We expect it drop gently from its maximum, around 4.2V, to the cutoff voltage programmed in the firmware at 3.3V.

```python
sts['bat'].plot(style='k')
ylabel('Battery voltage [mV]')
ax = sts.temp.plot(secondary_y=True, style='r')
ax.set_ylabel('Temperature [C]')
```

Great, there is a nice long plateau. We observe small dips when the temperature is too high. The ATmega1284p is rated for operations from -55C to 125C. We are largely withing this region, so what could it be ? From datasheet and reading from the internet, it seems the internal voltage reference might have some variability due to temperature which might explain it.

Radioactivity
-------------

This is the critical part. From the following graph, we can see that the CPM count starts to increase significantly when temperature is over 45C. We need to understand why, and also make sure that bGeigie normally do not reach this temperature. Having the temperature recorded along with the CPM data allows to calibrate this later and should be thus kept.

My first thought is that the heated up gas in the tube requires less energy to be ionized and create a pulse, leading to a higher sensitivity of the tube. This needs to be checked with Dan.

```python
rdd['CPM'].plot(style='k')
pd.rolling_mean(rdd['CPM'], 120).plot(style='y')
ylabel('CPM')
legend(('1 minute moving average', '2 hours moving average','Temperature'), loc='upper left')
ax = sts.temp.plot(secondary_y=True, style='r')
ax.set_ylabel('Temperature [C]')
```

GPS
---

Finally, it is interesting to observe the effect of temperature on the GPS coordinates. To do so, I compute the deviation from the mean GPS coordinate. Since the sensor was not moved during the testing period, the mean of the location should be close to the true position.

We observe that during temperature transient, when the temperature is changing abruptly, the accuracy of the positionning is affected. When the temperature becomes stable again, the accuracy is restored. This leads me to think that the GPS module incorporates some mechanism for temperature compensation. In addition, we observe that the extent of the deviation seems to be directly proportional to the temperature change.

Anyhow, the maximum error is only of a few hundred meters, which is not so bad.

```python
def distance(p1, p2, radius=6367445, unit='deg'):
    """
    Computes the distance between two (lat,lon) pairs.
    The optional argument radius can be given. It defaults to radius of earth in Lausanne, CH.
    """
    if unit is 'deg':
        lat1 = p1[0]*np.pi/180.
        lon1 = p1[1]*np.pi/180.
        lat2 = p2[0]*np.pi/180.
        lon2 = p2[1]*np.pi/180.
    else:
        lat1, lon1, lat2, lon2 = p1[0], p1[1], p2[0], p2[1]
        
    a = np.sin((lat1-lat2)/2.)**2 + np.cos(lat1)*np.cos(lat2)*np.sin((lon1-lon2)/2.)**2
    c = 2*np.arctan2(np.sqrt(a), np.sqrt(1-a))
    return radius*c

lat_i, lat_f = np.modf(rdd.Lat/100.)
lat = lat_i + lat_f*100./60.
lon_i, lon_f = np.modf(rdd.Lon/100.)
lon = lon_i + lon_f*100./60.

err = distance((lat, lon), (lat.mean(), lon.mean()))

plot(rdd.index, err, 'k')
ylabel('Deviation from mean [m]')
xlabel('Date')
#legend(('Latitude deviation','Longitude deviation'), loc='upper left')
ax = sts.temp.plot(secondary_y=True, style='r')
ax.set_ylabel('Temperature [C]')
```
